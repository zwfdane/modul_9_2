## Bezpośredni przełożony a ocena wydajności pracy 

Przyjrzymy się teraz zależności pomiędzy bezpośrednim przełożonym a oceną wydajności pracy pracownika. 
Pomocna będzie zbiorcza tabela zawierająca procentowy udział poszczególnych ocen dla każdego menedżera.

```python
# Zależność między bezpośrednim przełożonym a oceną wydajności
df_MP = pd.crosstab(df['ManagerName'],df['PerformanceScore'],normalize='index')
df_MP.head()
```

```python
# Rysujemy wykres dla odpowiednich odsetków ocen
axes = df_MP.plot.bar()
axes.legend(loc=(1.05,0.0))
```

U wszystkich menadżerów największy odsetek ocen stanowi "Fully Meets". W pięciu przypadkach drugi co do wielkości odsetek ocen stanowi "Exceeds". Również w pięciu przypadkach nie wystawiono ocen "PIP". Na podstwie wykresu można wnioskować, że istnieje zależność między ocenami a menedżerami.

Zauważmy, że analiza zależności liczbowych nie ma sensu: ID menager nie wnosi żadnej informacji ilościowej. Przykładowo z wykresu jointplot poniżej możemy wnioskować, że najwięcej wystawianych ocen to "Fully Meets". Jest on jednak zdecydowanie mniej czytelny niż wykres słupkowy powyżej.

```python
sns.jointplot(x='ManagerID',y='PerformanceScore',data=df)
```

## Źródła pozyskania pracownika a staż pracy 

Przyjrzymy się teraz zależności pomiędzy źródłem pozyskania pracownika a stażem w przedsiębiorstwie.
Najpierw wyznaczymy staż dla każdego pracownika oraz dodamy kolumnę staż.

```python
import datetime as dt

def count_seniority(row):

    if pd.isnull(row['DateofTermination']):
        end_date = dt.datetime(2022,8,19)
    else:
        end_date = row['DateofTermination']

    return (end_date - row['DateofHire'])/np.timedelta64(1,'Y')
```

```python
df['Seniority'] = df.apply(lambda row: count_seniority(row),axis=1)
```

Przedstwaimy teraz zalezność pomiędzy stażem pracy a źródłem pozyskania pracownika.

```python
sns.boxplot(x='Seniority',y='RecruitmentSource',data=df)
```

Średnio najdłuższy czas pracy mają pracownicy rekrutowani ze źródeł: "Other", "Bilboard", "Website Banner Ads".
